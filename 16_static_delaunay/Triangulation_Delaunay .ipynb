> **Триангуляция** заданного набора точек $-$ планарное разбиение плоскости
на фигуры, из которых одна является внешней неограниченной, а остальные –
треугольниками.

> **Глобальный критерий Делоне**: для любого треугольника нет точек из $S$, которые бы лежали внутри описанной около него окружности.

> **Триангуляция Делоне** $-$ триангуляция, удовлетворяющая глобальному критерию Делоне.

> **Локальный критерий Делоне для ребра**: для пары треугольников, которым принадлежит это ребро, выполняется критерий Делоне (то есть вершина, противолежащая ребру в одном треугольнике, не лежит в окружности, описанной вокруг другого, и наоборот).

### Лемма 1 (эквивалентность критерия Делоне для треугольников критерию для ребер)

> Если для всех рёбер выполняется локальный критерий Делоне, то выполняется и глобальный критерий Делоне.

$\triangleright$

<div style="padding-left:27px">
В одну сторону очевидно.<br>
Докажем в обратную сторону. Пусть для некоторого треугольника выполнен локальный, но не выполнен глобальный критерий. Т.е. существует точка $E$ такая, что лежит внутри описанной окружности и не имеет рёбер с вершинами треугольника. Зафиксируем эту точку и из всех таких треугольников рассмотрим тот, у которого $\angle BEC$ максимален, если $BC$ $-$ ближайшая к точке $E$ сторона. Пусть с таким $\Delta ABC$ соседствует $\Delta BCD$.
</div>
<img src="img/Bad_triangle.png" style="float: right; margin: 10px">
<div style="padding-left:27px">
Точка $E$ не может лежать внутри $\Delta BCD$, так как в этом случае имелись бы рёбра $EB$ и $EC$, и нарушался бы локальный критерий. Значит, она снаружи. Не уменьшая общности, будем считать, что $BD$ $-$ ближайшая к $E$ сторона $\Delta BCD$. Тогда выполнено, что $\angle BAC + \angle BEC > 180°, \angle BAC + \angle BDC < 180°$. Если точка $E$ лежит вне окружности, описанной около $\Delta BCD$, то $\angle BEC < \angle BDC$, что противоречит неравенствам выше. Значит, для $\Delta BCD$ также нарушается глобальный критерий. Из вышесказанного следует, что $\angle BED > \angle BEC$, где $BD$ $-$ ближайшая к $E$ сторона. 
</div>
<div style="padding-left:27px">
Очевидно, что $\angle BED$ больше, чем $\angle BEC$. При этом точка $E$ лежит в окружности, описанной вокруг $\Delta BDC$. Значит, при выборе треугольника нужно было взять не $\Delta ABC$, а $\Delta BDC$. Противоречие.
</div>

## Существование триангуляции Делоне и её построение

Далее необходимо показать существование триангуляции Делоне для произвольного набора точек. Для этого воспользуемся следующим методом. Поднимем все наши точки в просторанство большей размерности, преобразовав точки по такому закону: $(x, y) \to (x, y, x^2+y^2)$. Тем самым, мы спроецируем наши точки на параболоид $x^2 + y^2 = z$. Далее построим выпуклую оболочку полученных точек. Грани выпуклой оболочки будут соответствовать треугольникам в нашей триангуляции с некоторыми незначительными поправками.

### Лемма 2 (об окружности, спроецированной на параболоид)

> Все точки окружности, спроецированной на параболоид, находятся в одной плоскости. Точки, лежащие внутри окружности, будут лежать под этой плоскостью. Точки, лежащие вне окружности, будут лежать над плоскостью.

$\triangleright$

<div style="padding-left:27px">
Любая окружность в $R^2$ описывается уравнением $(x-a)^2+(y-b)^2=r^2$, где $(a, b)$ $-$ центр окружности, $r$ $-$ её радиус. 
Раскрываем скобки: $x^2-2ax+a^2+y^2-2by+b^2-r^2=0$.
Подставляем в уравнение параболоида $(x^2+y^2=z)$: $(-2a)x+(-2b)y+z+(a^2+b^2-r^2)=0$.
Получившееся уравнение является уравнением плоскости: $Ax+By+Cz+D=0$.
Это значит, что все точки, лежащие на одной окружности, при проецировании на параболоид будут лежать в одной плоскости.
</div>

<img src="img/TDD.png" style="padding-left:27px">

<div style="padding-left:27px">
Рассмотрим точки внутри этой окружности. Для них $(x-a)^2+(y-b)^2<r^2$. Подставляя в уравнение параболоида, получим   $Ax+By+Cz+D<0$, что означает, что точки, лежащие внутри окружности, будут находиться ниже плоскости спроецированной окружности. Аналогично для точек вне окружности.</div>

$\triangleleft$


Таким образом мы можем предложить примерный алгоритм построения триангуляции Делоне:

1. Перенести все точки на параболоид 
2. Построить выпуклую оболочку
3. Взять от неё lower convex hull (так как нужно убрать верхние грани выпуклой оболочки)
4. Перенести все точки и построенные ребра на плоскость


<h4>***Замечание!***</h4> Для того, чтобы убрать верхние грани, перейдем в однородные координаты, и введем бесконечно удаленную точку вверх (0, 0, 1, 0) и уберем все грани, смежные с ней, так как они будут бесконечными. Этот способ удобен тем, что, когда мы рассматриваем триангуляцию Делоне на плоскости, у нас возникает вопрос $-$ что делать с точками, лежащими вне выпуклой оболочки? Тогда мы говорим, что эти точки, называемые бесконечными, попадают в бесконечные треугольники, то есть все точки выпуклой оболочки будут соединены с этой бесконечной точкой, а значит и бесконечные грани также будут с ней соединены. Стоит отметить, что предикаты поворота и предкаты для флипов будут также корректны для бесконечных точек.

### Теорема о существовании триангуляции Делоне

> Триангуляция Делоне существует для любого множества точек.

<img src="img/parabol.png" width="300px" height="400px" style="float: right; margin: 10px">

$\triangleright$
<div style="padding-left:27px">
Все грани выпуклой оболочки окажутся внутри параболоида из-за его выпуклости. При этом точки лежат на параболоиде. Поэтому не найдётся точек, которые будут лежать за гранями выпуклой оболочки. То есть все точки, спроецированные на параболоид, будут принадлежать выпуклой оболочке. Внутри окружностей, описанных вокруг проекций треугольников выпуклой оболочки, не будет лежать никаких точек по лемме 2. Значит, для проекций треугольников выполнен критерий Делоне. Из этого следует, что получившаяся триангуляция есть триангуляция Делоне.
</div>
$\triangleleft$

## Сведение любой триангуляции к триангуляции Делоне

Для начала определим операцию **флип** (flip):

> **Флип (flip)** $-$ это замена одной диагонали на другую в четырёхугольнике, образованном смежными треугольниками.

<center><img src="img/flipm.png"></center> 

Будем называть ребро **хорошим**, если для него выполнен локальный критерий Делоне, и **плохим** в обратном случае.
Необходимо показать, что любая триангуляция сводится к триангуляции Делоне за конечное число флипов. Для этого докажем несколько лемм:


### Лемма 3 (о хорошем ребре)

> В любом четырёхугольнике, образованном смежными треугольниками, два ребра не могут быть одновременно плохими.

$\triangleright$

<img src="img/flip.png" style="float: right; margin: 10px">

<div style="padding-left:27px" >
Зафиксируем такой четырёхугольник $ABCD$. Пусть $AC$ и $BD$ это плохие рёбра. В окружности, описанной около $\Delta ABC$ содержится точка $D$, так как $AC$ не удовлетворяет локальному критерию Делоне. Значит $\angle ABC + \angle ADC > 180°$. Аналогично показывается, что $\angle BAD + \angle BCD > 180°$. В итоге сумма углов четырёхугольника $ABCD$ больше $360°$, что невозможно. Значит хотя бы одна диагональ является хорошим ребром и ,значит, можно флипнуть плохое ребро.
</div>

$\triangleleft$

### Предикат для флипа

Из леммы 2 мы знаем, что окружность, образованная тремя точками, и плоскость, построенная по этим же точкам, спроецированных на параболоид, взаимно однозначно соотносятся. Для проверки плохого ребра воспользуемся неравенстовм $Ax+By+Cz+D<0$ и перенесем три точки, образующие треугольник и содержащие плохое ребро, и точку для проверки (в нашем случае $x, y, z$) на параболоид, после чего проверим знак предиката:

$\begin{vmatrix} A_x & A_y & A_x^2 + A_y^2 & 1 \\ B_x & B_y & B_x^2 + B_y^2 & 1 \\ C_x & C_y & C_x^2 + C_y^2 & 1 \\ x & y & x^2 + y^2 & 1 \end{vmatrix} \ge 0$

Если это неравенство сохраняется, то, действительно, наше ребро плохое и его нужно флипнуть.

### Предикат для флипа в бесконечных гранях

Напомним, что точки, лежащие вне выпуклой оболочки, называются бесконечными, обладают координатами (0, 0, 1, 0) и лежат в бесконечных гранях. Тогда проверка на то, является ли хорошим ребро, инцидентное бесконечно удалённой точке, упрощается: 

$\begin{vmatrix} A_x & A_y & A_x^2 + A_y^2 & 1 \\ B_x & B_y & B_x^2 + B_y^2 & 1 \\ C_x & C_y & C_x^2 + C_y^2 & 1 \\ 0 & 0 & 1 & 0 \end{vmatrix} = \begin{vmatrix} A_x & A_y & 1 \\ B_x & B_y & 1 \\ C_x & C_y & 1 \end{vmatrix}$

То есть достаточно проверить поворот трёх остальных точек образованного двумя бесконечными треугольниками четырёхугольника.

<b>Замечание!</b>
<img src="img/inf.png" style="float: right;height: 200px; margin: 10px">
Стоит отметить, что предикат для флипа в бесконечных гранях фактически задает выпуклую оболочку. Например, на рисунке видно, что точка $B$ не принадлежит выпуклой оболочке, и это действительно можно проверить предикатом (ребро $AC$ будет хорошим).

### Лемма 4 (о флипе плохого ребра)

> Флип плохого ребра увеличивает объем многогранника.

$\triangleright$

<div style="padding-left:27px" >

Рассмотрим два таких смежных треугольника, что ребро между ними является плохим. Спроецируем их на параболоид. Четыре точки, принадлежащие смежным треугольникам, при проекции на параболоид образуют тетраэдр.<br>
Проведём через какой-нибудь из двух треугольников плоскость. Вершина, противолежащая основанию тетраэдра, являющегося этим треугольником, лежит ниже этой плоскости (так как не выполняется локальный критерий Делоне), то есть тетраэдр лежит ниже многогранника, образующегося при проекции всей триангуляции на параболоид.<br>
После флипа станет выполняться локальный критерий Делоне, то есть многогранник станет включать в себя тетраэдр. Поэтому после флипа плохого ребра объём многогранника увеличится на объём этого тетраэдра.
</div>

<img src="img/bad_edge.png" style="float: center; margin: 10px">

$\triangleleft$

### Лемма 5 (о сведении флипами)

> Любая триангуляция сводится к триангуляции Делоне за конечное число флипов.

$\triangleright$

<div style="padding-left:27px" >
Всего триангуляций заданного множества точек конечное число, и среди них есть триангуляция Делоне. Последовательность флипов плохих рёбер триангуляции образует такую последовательность триангуляций, что объем многогранника может только увеличиваться (по лемме 4). Эта последовательность конечна (при этом последней в последовательности является триангуляция Делоне), значит, число флипов, требуемых для достижения триангуляции Делоне, тоже конечно.
</div>

$\triangleleft$
