# Принадлежность точки не выпоклому многоугольнику

## Описание алгоритма
Главная идея состои в том, что можно провести луч из точки вертикально вверх и посчитать количество пересечений луча с ребрами многоугольника. 
Главная проблема - если луч пройдет через вершину многоугольника. Такой случай надо описать отдельно. Для этого, давайте считать, что ребро - интервал между вершинами и правая вершина (у которой координата $x$ больше). Тогда идею алгоритма можно описать формально:

## Алгоритм
* Мысленно проведем вертикальный вверх луч из данной точки.
* Посчитаем количество пересечений с ребрами многоугольника: 
    Переберем все ребра (a,b) по таким правилам ($a_x<b_x$): 
    * Если точка $p$ лежит на ребре, то точка лежит в многоугольнике,
    * Если у ребра $a_x$ = $b_x$, то пропускаем его,
    * Если у ребра $p_x=b_x$ и $b_y>p_y$, то увеличим счетчик на $1$
    * Если у ребра $p_x = a_x$, то пропускаем,
    * Если $a_x<p_x<b_x$ или $b_x<p_x<a_x$ и поворот точек $(a,b,p)$ левый, то увеличиваем счетчик пересечений на $1$,
* Если количество пересечений четно, то точка лежит вне многоугольника, иначе точка лежит внутри.

## Время работы
На каждое ребро тратится $O(1)$ времени, значит алгоритм работает за $O(n)$, где $n$ - количество вершин многоугольника.
